# Problem 3: Agriculture – Rainfall Adequacy Checker

## Problem Statement
Create a function that analyzes seasonal rainfall data.

### Rules
- Calculate **average rainfall**
- If **Average ≥ required level** → `Adequate Rainfall`
- Otherwise → `Inadequate Rainfall`

### Real-Time Use
- Climate monitoring systems  
- Irrigation planning tools  

### Hint
- Use **function**
- Use **loop and math operations**

```python
def rainfall_data_analyzer(rainfall_data):
  # Initial values and the given data
  total_sum = 0
  required_level = 65

# calculate total of rainfall using loop
  for i in rainfall_data:
    total_sum += i
  print("Total Rainfall: ", total_sum)

# Calculate the no of days
  length = len(rainfall_data)
  print("Number of days: ", length)

  # Calculate avg of the total rainfall
  avg = total_sum / length
  print("Average Rainfall: ", avg)

# Using conditional statement analyzes seasonal rainfall data.
  if avg >= required_level:
    print("Status: Adequate Rainfall")
  else:
    print("Status: Inadequate Rainfall")

# Take a data from the user
rainfall_data = list(map(int, input("Enter rainfall data separated by space: ").split()))

# Function call4
rainfall_data_analyzer(rainfall_data)
```

# Problem 10: Smart Electricity Bill Analyzer

## Problem Statement
You are building a smart electricity billing system for residential users.  
Create a function that calculates the electricity bill and classifies usage level based on units consumed.

### Billing Rules
- First 100 units → ₹3 per unit  
- Next 100 units (101–200) → ₹5 per unit  
- Above 200 units → ₹7 per unit  

### Usage Classification
- Total bill < ₹500 → `Low Usage`  
- ₹500 – ₹1500 → `Moderate Usage`  
- > ₹1500 → `High Usage`  

### Real-Time Use
- Smart meter systems  
- Electricity board billing software  
- Energy consumption analytics  

### Hint
- Use **function** to calculate bill  
- Use **conditional logic** for slabs  
- Use **return values** (bill and usage status)  

```python
def electricity_bill_analyzer(units):
  # Calculate the total bill based on slab rates
  if units <= 100:
    total_bill = units * 3
  elif units <= 200:
    total_bill = (100 * 3) + ((units - 100) * 5)
  else:
    total_bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

  # Determine usage status based on the total bill amount
  if total_bill < 500:
    usage_status = "Low Usage"
  elif total_bill <= 1500:
    usage_status = "Moderate Usage"
  else:
    usage_status = "High Usage"

  # Display the results
  print("Units Consumed:", units)
  print("Total Bill: ₹", total_bill)
  print("Usage Status:", usage_status)

# Ask the user to enter the number of units consumed
units = int(input("Enter the number of units consumed: "))

# Call the function to analyze the electricity bill
electricity_bill_analyzer(units)
```
