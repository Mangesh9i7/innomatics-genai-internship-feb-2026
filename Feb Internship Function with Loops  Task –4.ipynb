##  Problem 1: Mobile Recharge Validation System

###  Problem Statement
Create a function to validate a mobile recharge request.


### Rules:
- Recharge amount must be ≥ ₹50  
- Valid plans: 199, 299, 399, 599  
- User should be allowed to retry if invalid  


### Real-Time Use:
- Telecom recharge apps  
- Payment gateways  


### Hint:
- Use **function** for validation  
- Use **while loop** for retry  

```python
def recharge_validation():
    # Minimum recharge amount
    min_amount = 50
    # Valid recharge plans
    recharge_plans = [199, 299, 399, 599]

    # Using while loop for retry
    while True:
        try:
            amount = int(input("Enter your Recharge amount in INR: "))

            # Minimum recharge check
            if amount < min_amount:
                print("Sorry. Amount must be at least ₹50. Try again.")
                continue

            # Valid plan check
            if amount in recharge_plans:
                print(f"Congratulations! Recharge successful with plan ₹{amount}")
                break
            else:
                print("Sorry. Invalid plan. Available plans: 199, 299, 399, 599.")
                continue

        except ValueError:
            print("Invalid input. Please enter a number.")

# Run the function
recharge_validation()
```

##  Problem 3: Student Result Processing System

###  Problem Statement
Create a function that:  
Accepts a list of student marks  

Calculates average marks  

Prints Pass if average ≥ 50 else Fail  


###  Real-Time Use
- Online exam systems  
- Evaluation portals  


###  Hint
- Use **loop** to calculate total  
- Return result from function  

```python
def result():
  while True:
      try:

         # Take multiple marks input from the user in one line, separated by spaces
        marks = list(map(int, input("Enter your marks separated by spaces: ").split()))

        # Calculate total marks
        total_marks = sum(marks)

        if len(marks) == 0:
            print("No marks entered. Please enter valid marks.")
            continue

        # Calculate average marks
        avg_marks = total_marks / len(marks)

        # Cheak if avg marks is more than 50
        if avg_marks >= 50:
            print("Congratulations! You have passed", avg_marks)
            break
        else:
            print("Sorry, you have failed. Your average marks is", avg_marks)

       # Cheak if there is other input than integers
      except ValueError:
        print("Invalid input. Please enter valid marks.")


# Call the function
result()
```

##  Problem 4: Cab Fare Estimator with Retry Option

###  Problem Statement
Create a cab fare calculator:  
Base fare = ₹50  

₹12 per km  

Peak hour → 25% extra  

Allow retry using while loop  


###  Real-Time Use
- Ride-hailing apps  
- Transport systems  


###  Hint
- Function for fare calculation  
- While loop for retry  

```python
def cab_fare_calculator():
    base_price = 50
    per_km_price = 12

    while True:
        try:
            distance = float(input("Enter the distance traveled in km: "))
            peak_hour = input("Is it peak hour? (yes/no): ").strip().lower()

            # Calculate fare
            if distance <= 5:
                fare = base_price
            else:
                extra_distance = distance - 5
                fare = base_price + (extra_distance * per_km_price)

            # Apply peak hour surcharge
            if peak_hour == "yes":
                fare *= 1.25

            print("Your estimated cab fare is:", fare)
            break

        except ValueError:
            print("Invalid input. Please enter a number.")

cab_fare_calculator()
```

##  Problem 5: Employee Attendance Eligibility Checker

###  Problem Statement
Create a function that:  
Accepts attendance list ("P", "A")  

Calculates attendance percentage  

Returns Eligible if ≥ 75% else Not Eligible  


###  Real-Time Use
- HR payroll systems  
- Internship tracking  


###  Hint
- Use loop to count present days  
- Use return statement  

```python
def attendance_cheak():
  while True:             # Loop (though you break after one run)
    try:
      # Initial attendance list
      attendance = [
        "P","P","P","A","P","P","P",
        "A","P","P","P","A","P","P",
        "P","A","P","P","P","A","P",
        "P","P","A","P","P","P","A"
      ]

      # Ask user for today's attendance and convert to uppercase
      new_day = input("Enter your attendance (P/A): ").upper()

      # Add the new entry to the list
      attendance.append(new_day)

      total_dayes_of_month = 29

      # Count present days
      present_days = 0
      for day in attendance:
        if day == "P":
          present_days += 1

      # Calculate attendance percentage
      attendance_count = (present_days / total_dayes_of_month) * 100

      # Check eligibility
      if attendance_count >= 75:
        print("Congratulations! You are eligible for attendance.")
        break   # Exit loop
      else:
        print("Sorry, you are not eligible for attendance.")
        break   # Exit loop

    except ValueError:
      # This block would run if input conversion failed,
      # but here you’re not converting to int/float, so ValueError won’t occur.
      print("Invalid input. Please enter valid marks.")

attendance_cheak()   # Call the function
```

##  Problem 6: Password Strength Checker

###  Problem Statement
Create a function to check password strength.  

###  Rules:
Minimum length = 8  

Must contain at least one digit  

Must contain at least one special character (@#$)  


###  Real-Time Use
- Login systems  
- Security validation  


###  Hint
- Use loop over string  
- Use conditions inside function  


```python
def pass_strength_checker():
  while True:
      # Get password from user
      password = input("Enter your password: ")
      if len(password) < 8:
        print("Password must be at least 8 characters long.")
        continue
      if not any(char.isdigit() for char in password):
        print("Password must contain at least one digit.")
        continue
      if not any(char in "!@#$%^&*" for char in password):
        print("Password must contain at least one special character.  (@#$)")
        continue
      print("Password is strong.")
      break
pass_strength_checker()
```
