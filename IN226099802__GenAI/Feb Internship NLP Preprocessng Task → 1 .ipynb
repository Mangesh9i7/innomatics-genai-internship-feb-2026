# Task 2: Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):

    # Handle empty input string gracefully
    if not text:
        return [], ''

    # Convert to lowercase
    text = text.lower()

    # Remove URLs (e.g., http://example.com, www.example.com)
    text = re.sub(r'http\S+|www\S+|\S+\.com\S*', '', text)

    # Remove email-like patterns (e.g., user@domain.com)
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Normalize repeated characters (more than 2 occurrences become 1)
    # e.g., 'soooo goooood!!!' -> 'so good!!!'
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove punctuation and emojis (keep letters and spaces)
    # Emojis are generally non-alphanumeric, so this regex helps remove them.
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces and strip leading/trailing spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize the cleaned text
    words = text.split()

    # Remove very short tokens (length <= 2), but keep 'no' and 'not'
    filtered_words = [
        word for word in words
        if len(word) > 2 or word in {'no', 'not'}
    ]

    # Join the filtered words back into a cleaned sentence string
    cleaned_text = ' '.join(filtered_words)

    return filtered_words, cleaned_text

# Example usage for verification (optional, can be moved to Task 3)
# print(preprocess_text("WOW!!! Visit http://example.com NOW!!!"))
```

# Task 3: Stress Testing

```python
# Sample Input Sentences for stress testing
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("--- Stress Test Results ---")
for i, sentence in enumerate(sentences):
    tokens, cleaned_sentence = preprocess_text(sentence)

    print(f"\nSentence {i+1}:")
    print(f"  Original Text: '{sentence}'")
    print(f"  Cleaned Tokens: {tokens}")
    print(f"  Cleaned Sentence: '{cleaned_sentence}'")
    print("----------------------------------")
```

# Task 4: Token Analytics

```python
sentence_analytics = []

for i, sentence in enumerate(sentences):
    # Calculate original word count and length before any preprocessing
    original_word_count = len(sentence.split()) # Simple split for original count
    original_length = len(sentence)

    # Preprocess the sentence using the defined function
    tokens, cleaned_text = preprocess_text(sentence)

    # Calculate metrics after preprocessing
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    average_token_length = 0
    if total_tokens > 0:
        average_token_length = sum(len(token) for token in tokens) / total_tokens

    # Calculate 'noise' based on character reduction and 'meaningful' by token count
    cleaned_length = len(cleaned_text) # length of the cleaned string
    noise_level = original_length - cleaned_length # Characters removed as a proxy for noise
    meaningful_tokens_retained = total_tokens # Count of tokens remaining after cleaning

    sentence_analytics.append({
        'original_text': sentence,
        'cleaned_tokens': tokens,
        'total_tokens': total_tokens,
        'unique_tokens': unique_tokens,
        'average_token_length': average_token_length,
        'noise_level': noise_level,
        'meaningful_tokens_retained': meaningful_tokens_retained
    })

print("\n--- Token Analytics Results per Sentence ---\n")
for data in sentence_analytics:
    print(f"Original: '{data['original_text']}'")
    print(f"  Cleaned Tokens: {data['cleaned_tokens']}")
    print(f"  Total Tokens: {data['total_tokens']}")
    print(f"  Unique Tokens: {data['unique_tokens']}")
    print(f"  Average Token Length: {data['average_token_length']:.2f}")
    print(f"  Noise Level (Chars Removed): {data['noise_level']}")
    print(f"  Meaningful Tokens Retained: {data['meaningful_tokens_retained']}")
    print("----------------------------------")

# --- Analysis Questions ---
print("\n--- Analysis Questions ---\n")

# Which sentence had the most noise?
# Noise is defined as the reduction in string length (characters removed).
# Find the maximum noise level first
if sentence_analytics:
    max_noise_level = max(s['noise_level'] for s in sentence_analytics)
    # Find all sentences that have this maximum noise level
    most_noisy_sentences = [s for s in sentence_analytics if s['noise_level'] == max_noise_level]
    print(f"Which sentence had the most noise? (Max Noise Level: {max_noise_level} characters removed)")
    for s in most_noisy_sentences:
        print(f"  Answer: '{s['original_text']}'")
else:
    print("No sentences to analyze for noise.")

# Which sentence retained the most meaningful tokens after cleaning?
# Meaningful tokens are defined as the count of tokens remaining after preprocessing.
# Find the maximum number of meaningful tokens retained first
if sentence_analytics:
    max_meaningful_tokens = max(s['meaningful_tokens_retained'] for s in sentence_analytics)
    # Find all sentences that retained this maximum number of tokens
    most_meaningful_sentences = [s for s in sentence_analytics if s['meaningful_tokens_retained'] == max_meaningful_tokens]
    print(f"\nWhich sentence retained the most meaningful tokens after cleaning? (Max Retained Tokens: {max_meaningful_tokens})")
    for s in most_meaningful_sentences:
        print(f"  Answer: '{s['original_text']}'")
else:
    print("No sentences to analyze for meaningful tokens.")
```

# Task 5: Frequency Analysis

```python
from collections import Counter

# Combine all cleaned tokens from all sentences processed in Task 4
all_tokens = []
for data in sentence_analytics:
    all_tokens.extend(data['cleaned_tokens'])

# Perform frequency analysis using Counter
token_frequencies = Counter(all_tokens)

print("\n--- Token Frequency Analysis ---\n")

if token_frequencies:
    # Identify Top 10 most frequent words
    print("Top 10 Most Frequent Words:")
    for word, count in token_frequencies.most_common(10):
        print(f"  '{word}': {count}")

    # Identify Top 5 least frequent words
    # To get the least frequent, sort all items by count (ascending) and then by word (for stable order)
    sorted_by_frequency = sorted(token_frequencies.items(), key=lambda item: (item[1], item[0]))

    print("\nTop 5 Least Frequent Words:")
    # Ensure we don't try to print more unique words than exist if the list is very short
    num_least_frequent = min(5, len(sorted_by_frequency))
    for word, count in sorted_by_frequency[:num_least_frequent]:
        print(f"  '{word}': {count}")
else:
    print("No tokens found for frequency analysis.")
```

# Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    """
    Processes a list of text strings using the preprocess_text function
    and aggregates the results.

    Args:
        text_list (list): A list of raw text strings.

    Returns:
        dict: A dictionary containing:
            - 'tokens' (list): A combined list of all cleaned tokens from all input texts.
            - 'clean_sentences' (list): A list of cleaned sentence strings, corresponding to each input text.
    """
    all_tokens_combined = []
    cleaned_sentences_list = []

    for text in text_list:
        tokens, cleaned_sentence = preprocess_text(text)
        all_tokens_combined.extend(tokens)
        cleaned_sentences_list.append(cleaned_sentence)

    return {
        "tokens": all_tokens_combined,
        "clean_sentences": cleaned_sentences_list
    }

# Demonstrate the full_pipeline function
print("\n--- Full Pipeline Demonstration ---")
new_texts = [
    "Hello world, this is a test!",
    "123 abc DEF GHI jkl.com...",
    "Another example with noooooise!!!"
]

pipeline_output = full_pipeline(new_texts)
print("Pipeline Output:", pipeline_output)
print("Input Texts:", new_texts)
```
