## Task 4: ChatPromptTemplate System

```python
from langchain_core.prompts import ChatPromptTemplate

# Build the chat-based prompt system
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are an expert {role}."),
    ("human", "My topic is {topic}.")
])

# Test with different roles
roles = ["teacher", "interviewer", "motivator"]
topic = "Neural Networks"

for role in roles:
    messages = chat_template.format_messages(role=role, topic=topic)
    print(f"Role: {role.capitalize()}")
    # Print the resulting message objects
    for msg in messages:
        print(f"  {msg.type.capitalize()}: {msg.content}")
    print()
```

## Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # Assign defaults if invalid
    final_audience = audience if audience in valid_audiences else "beginner"
    final_tone = tone if tone in valid_tones else "casual"

    # Optional: Alert the user if defaults were applied
    if audience not in valid_audiences or tone not in valid_tones:
         print(f"[Warning] Invalid inputs. Defaulting to Audience: '{final_audience}', Tone: '{final_tone}'")

    return final_audience, final_tone

# Test valid and invalid cases
print("Valid input:", validate_inputs("expert", "formal"))
print("Invalid input:", validate_inputs("alien", "angry"))
```

## Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):
    # 1. Validate Inputs
    val_audience, val_tone = validate_inputs(audience, tone)

    # 2. Define Styles Dictionary
    style_templates = {
        "teaching": "Explain {topic} for {audience}s in a {tone} tone, clearly step by step.",
        "interview": "Ask 3 {tone} questions about {topic} suitable for an {audience} audience.",
        "storytelling": "Explain {topic} as a {tone} story meant for {audience}s."
    }

    # 3. Select Template (fallback to 'teaching' if style is unrecognized)
    selected_template_string = style_templates.get(style.lower(), style_templates["teaching"])

    # 4. Generate LangChain PromptTemplate
    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=selected_template_string
    )

    # 5. Return Formatted Output
    return prompt.format(topic=topic, audience=val_audience, tone=val_tone)

# Example Output
final_prompt = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)

print("Generated Prompt:\n", final_prompt)
```
