#  1. User Login Check

### **Problem Statement**

* **Authorized Password:** `1234`

### **Requirements**
1.  **Input:** Accept a username and password.
2.  **Logic:** Compare input against the authorized credentials.
3.  **Output:**

#  2. Pass / Fail Analyzer

### **Problem Statement**
Analyze student results.

* **Data Set:** `marks = [45, 78, 90, 33, 60]`
* **Passing Threshold:** 50 or higher

### **Requirements**
1.  A student passes if marks â‰¥ 50
2.  Count the total number of pass students
3.  Count the total number of fail students
4.  Print the final result clearly

### **Real-World Application:**
 Academic evaluation systems


```python
# Given data
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Logic to cheak student is Pass or Fail using for loop and conditional statement
for mark in marks:
    if mark >= 50:
        pass_count = pass_count + 1  # Increment pass counter
    else:
        fail_count = fail_count + 1  # Increment fail counter

# --- Output ---
print("ðŸ“ CLASS RESULTS")
print(f" Passed:      {pass_count}")
print(f" Failed:      {fail_count}")
```

# 3. Simple Data Cleaner

### **Problem Statement**
Clean and standardize user names

* **Raw Data:** `names = [" Alice ", "bob", " CHARLIE "]`

### **Requirements**
1.  **Remove extra spaces**
2.  **Convert all names to lowercase**
3.  **Print the cleaned list**

### **Real-World Application**
  Data preprocessing before analysis

```python
# Given raw data
names = [" Alice ", "bob", " CHARLIE "]

# Create empty lists to store the different stages
data_stripped = []
data_lower = []
cleaned_names = []

# Logic for Cleaning data
for name in names:
    # .strip() used removes spaces from start and end
    s_name = name.strip()
    data_stripped.append(s_name)

    # .lower() used converts text to lowercase
    l_name = name.lower()
    data_lower.append(l_name)

    # Final Step: We do both to get the perfect clean name
    final_name = name.strip().lower()
    cleaned_names.append(final_name)

# --- Output ---
print("1] Data without extra space:", data_stripped)
print("2] Data in lowercase:       ", data_lower)
print("3] Cleaned Data:            ", cleaned_names)
```

#  4. Message Length Analyzer

### **Problem Statement**
Analyze message sizes.
* **Messages:** `["Hi", "Welcome to the platform", "OK"]`

### **Requirements**
1.  **Print the length of each message**
2.  **Flag messages longer than 10 characters**

### **Real-World Application**
Text filtering and validation systems

```python
# Given data
messages = ["Hi", "Welcome to the platform", "OK"]

# Logic for Analysis
for msg in messages:
    length = len(msg)

    # 2. Check if the message is too long (more than 10)
    if length > 10:
        print(f"Message: '{msg}'")
        print(f"Length:  {length} Message is Greater than 10 character")
    else:
        print(f"Message: '{msg}'")
        print(f"Length:  {length} Message is Less than 10 characters")

    print("-" *90)
```

# 5. Error Message Detector

### **Problem Statement**
Detect error messages from system logs.
* **Log :** `["INFO", "ERROR", "WARNING", "ERROR"]`

### **Requirements**
1.  **Count the number of "ERROR" entries**
2.  **Print the total error count**


### **Real-World Application**
 Monitoring and log analysis systems

```python
# Give data
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Initialize a counter to start at zero
error_count = 0

#Logic for Analysis
for log in logs:
    if log == "ERROR":
        error_count = error_count + 1
        print(f"Found: {log} ")
    else:
        print(f"Found: {log} ")

print("-" * 30)
print(f" Total Errors Found: {error_count}")
print("-" * 30)
```
