#  Problem 1: Smart Light Controller

##  Problem Statement
Create a class to control a smart light that can be turned **ON** or **OFF** and display its status.

###  Real-Time Use
Smart home automation systems

###  Sample Input
- Light Name: Bedroom Light  
- Action: ON  

```python
class Smartlight:
    def __init__(self, name):
        self.name = name

    # Method to check and display the status of the light
    def status_on(self, action):
        if action == "on":
            print(self.name, "is", action)
        else:
            print(self.name, "is", action)

# Sampal input
light = Smartlight("Bedroom Light")
# Action on
light.status_on("on")
```

#  Problem 2: Employee ID Card System

##  Problem Statement
Create a class that stores **employee name, ID, and department** and displays ID card details.

###  Real-Time Use
Corporate HR systems

###  Sample Input
- Employee Name: Rahul  
- Employee ID: EMP102  
- Department: IT  



```python
class Employee:
    def __init__(self, employee_name, employee_id, department):
        # Initialize employee details
        self.employee_name = employee_name
        self.employee_id = employee_id
        self.department = department

    # Function to display employee details
    def id_card_display(self):
        print("Employee ID Card")
        print("Employee Name:", self.employee_name.upper())
        print("Employee ID:", self.employee_id.upper())
        print("Department:", self.department.upper())
        print()


# ðŸ“¥ Sample Input
employee1 = Employee("Mangesh", "EMP32", "aiml")
employee2 = Employee("Atul", "EMP57", "cs")

employee1.id_card_display()
employee2.id_card_display()
```

# Problem 3: Mobile Contact Record

## Problem Statement
Create a class that stores **contact name and phone number** and displays contact information.

### Real-Time Use
Mobile contact applications

### Sample Input
- Contact Name: Anita  
- Phone Number: 9876543210  

```python
class Contact:
    def __init__(self, name, phone_no):
        self.name = name
        self.phone_no = phone_no
   # Function to display saved contact info
    def contact_info(self):
        print("Contact Save")
        print("Name:", self.name)
        print("Phone NO:", self.phone_no)
        print()

# Sample input
contact1 = Contact("Mangesh", "9885569578")
contact2 = Contact("Atul", "9876543210")

contact1.contact_info()
contact2.contact_info()
```

# Problem 4: Product Price Tag Generator

## Problem Statement
Create a class that stores **product name and price** and prints a formatted price tag.

### Real-Time Use
Retail billing systems

### Sample Input
- Product Name: Headphones  
- Price: 2499  

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    # Function to display Product info
    def price_tag(self):
        print("Product Price Tag")
        print("Product Name:", self.name)
        print("Price", self.price)
        print()

# Sample input
product1 = Product("Headphones", "2499")
product2 = Product("Powerbank", "1499")

product1.price_tag()
product2.price_tag()
```

# Problem 5: Movie Rating Display System

## Problem Statement
Create a class that stores **movie name and rating** and displays movie details.

### Real-Time Use
Streaming platforms

### Sample Input
- Movie Name: Inception  
- Rating: 4.8  

```python
class Rating:
  def __init__(self, name, rating):
    self.name = name
    self.rating = rating

   # Function to display movie info
  def movie_details(self):
    print("Movie Rating")
    print("Movie Name:", self.name)
    print("Rating:", self.rating, "/5")
    print()


# Sample output
rating1 = Rating("Inception", "4.8")

rating1.movie_details()
```
